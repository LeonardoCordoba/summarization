```python
notas = [nota_1, nota_2, nota_3, nota_4, nota_5]
```

```python
def generate_text(text, model, tokenizer):
    inputs = tokenizer([text], padding="max_length", truncation=True, max_length=512, return_tensors="pt")
    input_ids = inputs.input_ids.to("cuda")
    attention_mask = inputs.attention_mask.to("cuda")

    outputs = model.generate(input_ids, attention_mask=attention_mask, max_length=64)

    # all special tokens including will be removed
    output_str = tokenizer.batch_decode(outputs, skip_special_tokens=True)

    return output_str
```

```python
resultados = []
for nota in notas:
  mt5 = generate_text(nota, mt5_model, mt5_tokenizer)
  b2b = generate_text(nota, b2b_model, b2b_tokenizer)
  resultados.append({"mt5":mt5, "b2b":b2b})
```

```python
import tqdm
```

```python
mt5_length = []
b2b_length = []
mt5_generated = []
b2b_generated = []
base_text = []
gold = []

for i in tqdm.tqdm(test):
    mt5_text = generate_text(i["text"], mt5_model, mt5_tokenizer)
    mt5_generated.append(mt5_text)
    b2b_text = generate_text(i["text"], b2b_model, b2b_tokenizer)
    b2b_generated.append(b2b_text)
    base_text.append(i["text"])
    gold.append(i["summary"])

    mt5_length.append(len(mt5_text[0].split(" ")))
    b2b_length.append(len(b2b_text[0].split(" ")))
```
